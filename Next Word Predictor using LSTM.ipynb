```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tk=Tokenizer()
```

```python
tk.fit_on_texts([data])
```

```python
input_seq=[]
for sentence in data.split('\n'):
  tks=tk.texts_to_sequences([sentence])[0]

  for i in range(1, len(tks)):
    input_seq.append(tks[:i+1])

```

```python
max([len(x) for x in input_seq])
```

```python
# Now lets do padding
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
pd=pad_sequences(input_seq, maxlen=18, padding='pre')
```

```python
pd
```

```python
x=pd[:,:-1]
y=pd[:,-1]
```

```python
y
```

```python
x.shape
```

```python
from keras.utils import to_categorical
y=to_categorical(y, num_classes=1153)
y
```

```python
from keras import Sequential
from keras.layers import *
```

```python
model = Sequential([
    Input(shape=(17,)),
    Embedding(1153, 100),
    LSTM(150),
    Dense(1153, activation='softmax')
])
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x,y, epochs=100)
```

```python
import numpy as np
```

```python
text=input('Enter a word: ')
for i in range(10):

#Tokenizing word
  tkt=tk.texts_to_sequences([text])[0]
#Padding
  pdtk=pad_sequences([tkt], maxlen=17, padding='pre')

  pos=np.argmax(model.predict(pdtk))

  for word,index in tk.word_index.items():
    if index == pos:
      text=text + " " + word
      print(text)
```
